```python
import numpy as np
```

```python
import pandas as pd
```

```python
import pickle
```

```python
df=pd.read_csv('C:/Users/Asus/Untitled Folder 1/final_df_3.csv')
```

```python
X=df.drop('pIC50',axis=1)
```

```python
X
```

```python
y=df['pIC50']
```

```python
X.shape
```

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
def remove_low_variance(input_data, threshold=0.1):
    selection = VarianceThreshold(threshold)
    selection.fit(input_data)
    return input_data[input_data.columns[selection.get_support(indices=True)]]
```

```python
X=remove_low_variance(X,threshold=0.1)
```

```python
X.to_csv('Slimmed Descriptors.csv',index=False)
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
np.random.seed(100)
model=RandomForestRegressor(n_estimators=100)
model.fit(X_train,y_train)
```

```python
score=model.score(X_test,y_test)
score
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_squared_error,r2_score,accuracy_score
ac=mean_squared_error(y_test,y_pred)
print(f"Mean Squared Error: {ac:.2f}")
error=y_pred-y_test
print('Average absolute error:', round(np.mean(error), 2), 'degrees.')
mape = np.mean(100 * (error / y_test))
accuracy = 100 - mape
print('Accuracy:', round(accuracy, 2), '%.')
```

```python
pickle.dump(model,open('Model.pkl','wb'))
```
