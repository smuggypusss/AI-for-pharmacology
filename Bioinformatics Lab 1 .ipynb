# Info
##### 1)action_type: Descrbes the type of action the molecule has on the target(eg. Inhibitor, agonist)
##### 2)activity_comment: Any additional comments about the activity measurement
##### 3)activity_properties: Contains additional properties related to the activity, provided as a list. 
##### 4)assay_description: A brief description of the assay. In this data, it describes in vitro inhibitory concentration against SARS(this means )
##### 5)assay_type: Typically denoted by a single letter. B for biochemical assay
##### 6) assay_variant_accession: if the assay invloves a specific variant of the protein, this columns provides the accesion number.
##### 7) assay_variant_mutation: Any specific mutations in the protein used in the assay.
##### 8)bao_format: BioAssay Ontology format identifier, which classifies the assay format.
##### 9)bao_label: Label describing the BAO format. Here, it's "single protein format".
##### 10)canonical_smiles: A textual representation of the chemical structure of the molecule using the Simplified Molecular Input Line Entry System (SMILES) notation.
##### 11)data_validity_comment: Comments about the validity of the data. It's None here.
##### 12)data_validity_description: Description of data validity, which is also None in your data.
##### 13)document_chembl_id: A unique identifier for the document or publication where the data was reported.
##### 14)document_journal: The name of the journal where the data was published.
##### 15)document_year: The year the data was published.
##### 16)ligand_efficiency: Contains various ligand efficiency metrics like binding efficiency index (BEI), ligand-lipophilicity efficiency (LLE), etc. Here it's given as a dictionary.
##### 17)molecule_chembl_id: A unique identifier for the molecule in the Chembl database.
##### 18)molecule_pref_name: The preferred name for the molecule, if available. It's None here.
##### 19)parent_molecule_chembl_id: If the molecule is a derivative of another molecule, this column gives the Chembl ID of the parent molecule.
##### 20)pchembl_value: A standardized bioactivity value (pIC50, pKi, etc.).
##### 21)potential_duplicate: Indicates whether this record might be a duplicate (0 means no, 1 means yes).
##### 22)qudt_units: Quantities, Units, Dimensions and Types (QUDT) units, providing a URL linking to the units of measurement.
##### 23)record_id: Another unique identifier for the record.
##### 24)relation: The relational operator used in the activity measurement (e.g., '=', '<', '>').
##### 25)src_id: Source identifier indicating the origin of the data.
##### 26)standard_flag: Indicates whether the activity data is standardized (1 means yes, 0 means no).
##### 27)standard_relation: The relational operator for the standardized data (e.g., '=', '<', '>').
##### 28)standard_text_value: The textual representation of the standardized activity value. It's None here.
##### 29)standard_type: The type of activity measurement standardized (e.g., IC50).
##### 30)standard_units: Units of the standardized activity measurement (e.g., nM, uM).
##### 31)standard_upper_value: Upper bound of the standardized activity value, if available. It's None here.
##### 32)standard_value: The standardized activity value.
##### 33)target_chembl_id: A unique identifier for the target protein in the Chembl database.
##### 34)target_organism: The organism from which the target protein is derived.
##### 35)target_pref_name: The preferred name for the target protein.
##### 36)target_tax_id: The taxonomy identifier for the target organism.
##### 37)text_value: Original textual representation of the activity value. It's None here.
##### 38)toid: Target identifier, used in target ontology. It's None here.
##### 39)type: Type of activity measurement. In your data, it's IC50 (the concentration of the drug that inhibits the biological activity of the target by 50%).
##### 40)units: Units of the activity measurement (e.g., nM, uM).
##### 41)uo_units: Units in the Uniform Ontology (UO) format.
##### 42)upper_value: Upper bound of the activity measurement, if available. It's None here.
##### 43)value: The activity value.

###### The term IC50 stands for half maximal inhibitory concentration which is used to indicates the effectiveness of the substance in inhibiting a specific substance.
###### IC50 is the concentration of a compound at which 50% of its maximum inhibitory effect is observed.
###### Here it indicates the concentration of a compund needed to inhibit the activity of SARS coronavirus 3C- like proteinase enzyme by 50% in vitro.
###### Lower IC50 values have higher potency.

# Data Pre-processing of the bioactivity data

##### We will now segregate the different drugs based on their IC50 value indicated by Standard_value

```python
bioactivity_class=[]
for i in df.standard_value:
    if float(i)>= 10000:
        bioactivity_class.append('Inactive')
    elif float(i)<=1000:
        bioactivity_class.append('Active')
    else:
        bioactivity_class.append('Intermediate')
```

```python
df2=df
```

```python
selection = ['molecule_chembl_id','canonical_smiles','standard_value']
df3=df2[selection]
```

```python
df3=pd.concat([df3,pd.Series(bioactivity_class)],axis=1)
```

```python
df3.to_csv('bioactivity_preprocessed_data.csv',index=False)
```
